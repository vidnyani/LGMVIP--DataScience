Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

Reading the Dataset

```python
df = pd.read_csv(r'globalterrorismdb_0718dist.csv',encoding='ISO-8859-1', engine="python", error_bad_lines=False)
```

Data Visualization

Terrorist activities by year

```python
count =  df['iyear'].value_counts()
year = count.index.sort_values(ascending  = True)
val = []
for i in year:
    val.append(count[i])
```

```python
plt.figure(figsize = (20,10))
plt.bar(year, val, tick_label = year, color = 'red')
plt.xticks(rotation  = 90)
plt.title('Number of Terrorists Activities Per Year', fontsize=28)  
plt.xlabel('Year', fontsize = 20, labelpad = 32)
plt.ylabel('Number of terrorists attacks', fontsize = 18, labelpad = 30)
plt.grid(False)
plt.show()
```

Terrorist attacks by region

```python
pd.crosstab(df.iyear, df.region_txt).plot(kind='area',stacked=False,figsize=(15,6))
plt.title('Terrorist Activities By Region In Each Year',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xlabel("Year",fontsize=15)
plt.show()
```

Number of attacks on countries

```python
plt.subplots(figsize=(8,10))
sns.barplot(df['country_txt'].value_counts()[:16].values,df['country_txt'].value_counts()[:16].index,palette='Paired')
plt.title('Number of Attacks on Countries')
plt.ylabel('Countries')
plt.xlabel('Range')
plt.show()
```

Types of Attacks

```python
attack_type = df.attacktype1_txt.value_counts()[:15]
attack_type
```

```python
plt.subplots(figsize=(12,6))
sns.barplot(attack_type.index,attack_type.values,palette="rocket")
plt.title('Attack Types')
plt.xlabel('Types of attacks')
plt.ylabel('Number of attacks')
plt.show()
```

Most attacked country

```python
print("Most Attacked Country:",df["country_txt"].value_counts().idxmax())
```

Number of people killed by terror attacks

```python
killedpeople = df.loc[:,'nkill']
print('Number of people killed by terror attack:', int(sum(killedpeople.dropna())))
```

Gang with most number of attacks

```python
print("Gang with most number of attacks:",df['gname'].value_counts().index[1])
```

Most frequently used attack type

```python
print("Most frequently used attack type:",df['attacktype1_txt'].value_counts().idxmax())
```
